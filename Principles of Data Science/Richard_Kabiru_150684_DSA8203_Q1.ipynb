### EDA

```python
# Plot a Correlation heatmap between the variables
plt.figure(figsize=(10,10))
sns.heatmap(eval.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Plot a pairplot between the variables
sns.pairplot(eval)
```

```python
# Use the describe method to get the summary statistics of the dataset
eval.describe()
```

Describe was used to get the statistics of each column in the dataset. The count of the rows is 283 with categorical columns having values between 1 and 4

Pairplot was used to check for the distribution of data points depending on the different columns. The data mainly has categorical varaibles.

The correlation heat map was used to determine the correlation between the different columns. The columns were noted to have strong positive correlations.

```python
# Matplotlib plot a barplot to show the Best performing team from the which room column depending on their rating
plt.figure(figsize=(10,10))
sns.barplot(x='Which Room', y='Rating', data=eval, palette='coolwarm', order=eval.groupby('Which Room')['Rating'].mean().sort_values(ascending=False).index)
plt.title('Best performing group depending on their rating')
plt.show()
```

Group 6 had the best performing rating

```python
# Matplotlib plot a barplot to show the Best performing Rator from the RatorID column depending on their rating
plt.figure(figsize=(10,10))
sns.barplot(x='RatorID', y='Rating', data=eval, palette='coolwarm', order=eval.groupby('RatorID')['Rating'].mean().sort_values(ascending=False).index)
plt.title('Best performing Rator depending on their rating')
# rotate the xticks
plt.xticks(rotation=90)
plt.show()
```

Rator 44 is the best performing rator
