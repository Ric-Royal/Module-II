```python
# identify categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print(categorical_columns)
# one hot encode the categorical columns
df = pd.get_dummies(df, columns=categorical_columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Convert dataset to numpy array
data_1 = np.array(df)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform data using MinMaxScaler
data = scaler.fit_transform(data_1)
```

```python
from sklearn.random_projection import GaussianRandomProjection

# Create the random projection object
transformer = GaussianRandomProjection(n_components=1000)

# Fit and transform the data
X_new = transformer.fit_transform(data)
```

```python
print(X_new)
```

```python
# convert the output to a dataframe
data_grp_df = pd.DataFrame(X_new, columns=[f'GRP{i+1}' for i in range(1000)])
```

```python
data_grp_df.head()
```

```python
# show the most relevant columns
most_relevant_cols = transformer.components_.argsort()[:, -5:][::-1]
plt.plot(most_relevant_cols)
plt.xlabel('Index')
plt.ylabel('Column')
plt.title('Most Relevant Columns')
plt.show()
```

```python
# show the columns with the most input
most_important_cols = transformer.components_.argsort()[:,-1:][::-1]
plt.plot(most_important_cols)
plt.xlabel('Index')
plt.ylabel('Column')
plt.title('Most Important Columns')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Fit a PCA model to the transformed data
pca = PCA()
pca.fit(X_new)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(explained_variance)
```
